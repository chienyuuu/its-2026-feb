```python
# % 字串格式化
# print(項目 % (參數列))

'''
常用的參數
%d -> 整數
%s -> 字串
%f -> 浮點數
%% -> %
%e / %E -> 科學符號
'''

print("小明的成績是80分")
print("%s的成績是%d分" % ('小明', 80))

print("小明成績是班上 %d %%" % (90))
```

```python
# % 字串格式化
# 精確的控制輸出的位置 -> 整齊排列

# %5d：固定印出5個字元，少於5個就在左邊補上空白字元
print(123)
print("%5d" % (123))
print("%5d" % (12345))
print("%5d" % (1234567))

# %8.2f：固定印出8個字元（包含小數點），小數點後固定印出兩個字元
# 整數不足 -> 左邊補空白
# 小數不足 -> 右邊補 0
print("%8d" % (123))
print("%8.2f" % (123.4567))
print("%8.2f" % (123.4))
```

```python
'''
練習：
國稅局的資料，整齊列出這三年的各種稅收
"2017", 98.34, 90.2, 104.79
"2016", 83, 110.5, 82.45
"2015", 98, 79.32, 102
'''

print("%4s %6.2f  %6.2f  %6.2f" % ("2017", 98.34, 90.2, 104.79))
print("%4s %6.2f  %6.2f  %6.2f" % ("2016", 83, 110.5, 82.45))
print("%4s %6.2f  %6.2f  %6.2f" % ("2015", 98, 79.32, 102))
```

```python
# format 字串格式化
# 用大括號 {} 表示參數的位置
# print(字串.format(參數列))

print('{}的成績是{}分'.format('小明', 80))

# 多個參數，可在大括號放順序數字（從 0 開始）
print('{0}的成績是{1}分'.format('小明', 80))
print('考{1:3}的人是{0:4}'.format('小明', 80))
```

```python
# format 字串格式化

print('{0:4s}的成績是{1:3d}分'.format('amy', 80))
print('{0:4s}的成績是{1:3d}分'.format('alan', 100))

print('{0:4}的成績是{1:3}分'.format('alan', 100))
print('{:4}的成績是{:3}分'.format('alan', 100))
```

```python
# 判斷式

# 單向判斷式：只有 if
'''
練習：密碼輸入判斷
密碼是 1234
若輸入正確 則輸出歡迎光臨
'''

# 雙向判斷式：if ... else ...
'''
練習：密碼輸入判斷
密碼是 1234
若輸入正確 則輸出歡迎光臨
若輸入錯誤 則輸出密碼錯誤
'''

# 多向判斷式：if ... elif ... (else) ...
# 擇一選擇

# x 只有1~4這幾種可能
x = int(input())
if x == 1: print('輸入的是1')
elif x == 2: print('輸入的是2')
elif x == 3: print('輸入的是3')
else: print('輸入的是4')

'''
練習： 判斷成績等第
使用者輸入成績
90以上優等、80以上甲等、70以上乙等、60以上丙等、不到60丁等
'''

# 巢狀判斷式 nested：判斷式裡面包含判斷式
# 沒有層數限制
# 太多層會降低程式可讀性、比較難 debug

'''
練習：
a, b, c 是三個不同的正整數
用巢狀判斷式找出最大值
hint: 兩兩比較
'''
a = int(input())
b = int(input())
c = int(input())

# 先比較 a 和 b
if a > b: # a 比 b 大
  # a 去跟 c 比較
  if a > c: print("a")
  else: print("c")
else: # b 比 a 大
  # b 去跟 c 比較
  if b > c: print("b")
  else: print("c")
```

```python
# 判斷以下兩段程式有何不同

if x == 1: print(1)
elif x % 1 == 0: print(x)

if x == 1: print(1)
if x % 1 == 0: print(x)
```
