```python
'''
ch3 複習

1. % 字串格式化（固定向右對齊）：
  - %d, %f, %s, %%, %e
  - %4d：固定印出 4 個字元
  - %8.2f：固定印出 8 個字元（包含小數點），小數點後的位數 2 位
2. format 字串格式化（字串向左對齊、數字向右對齊）：
  - 大括號表示參數位置
  - 可以在大括號放上順序數字
3. 判斷式（單向、雙向、多向、巢狀）
'''

# print(項目 % (參數列))
print("%s 的成績是 %d" % ('amy', 90))
print("%4d" % (123))

print("%8.2f" % (123.4567))
print("%8.2f" % (123.4))


# print(字串.format(參數列))
print('{}的成績是{}分'.format('小明', 80))

print('{0}的成績是{1}分'.format('小明', 80))
print('考{1}的人是{0}'.format('小明', 80))

print('{0:4s}的成績是{1:3d}分'.format('小明', 80))
print('{0:4}的成績是{1:3}分'.format('小明', 80))
print('{:4}的成績是{:3}分'.format('小明', 80))
```

```python
# range

# 單一參數
# 數列變數 = range(整數值)
# 產生 0 至 (整數值 - 1) 的串列

list1 = range(5)
print(list(list1))
```

```python
# 兩個參數

# 數列變數 = range(起始值, 終止值)
# 產生 (起始值) 至 (終止值 - 1) 的串列

list2 = range(3, 8)
print(list(list2))

list3 = range(-5, 5)
print(list(list3))

'''
若起始值 >= 終止值，則產生空串列 []
'''
```

```python
# 三個參數

# 數列變數 = range(起始值, 終止值, 間隔值)
# 產生 (起始值) 開始，每次遞增 (間隔值)，直到 (終止值 - 1) 的串列

list4 = range(3, 8, 1)
print(list(list4))

list5 = range(3, 8, 2)
print(list(list5))

list6 = range(8, 3, -1)
print(list(list6))

list7 = range(8, 3, -2)
print(list(list7))
```

```python
'''
井字直角三角形
輸出格式：

外層 1 次，內層 1 次：#
外層 2 次，內層 2 次：##
外層 3 次，內層 3 次：###
外層 4 次，內層 4 次：####
外層 5 次，內層 5 次：#####
'''

for i in range(1, 6): # 控制層數 1 ~ 5
  print(f'外層 {i} 次，內層 {i} 次：', end='')
  for j in range(1, i+1): # 控制井字號數目
    print("#", end='')
  print()
```

```python
# while 迴圈：條件迴圈，未知次數

# while 條件式:
#   程式區塊

total = 0
i = 0
while i <= 10:
  total += i
  i += 1
print(total)
print(i)

'''
無窮迴圈
不小心寫出 -> control + c 跳出迴圈
'''
```

```python
'''
計算階層

1! = 1
2! = 1*2 = 2
3! = 1*2*3 = 6
'''

i = 1
ans = 1
n = int(input())
while i <= n:
  ans = ans * i
  i = i + 1

print(ans)
```

```python
'''
練習

使用者輸入一個正整數
用 while 迴圈印出 1 ~ 該正整數的「所有偶數」
'''
```

```python
# break：強制離開迴圈

# 找最小公倍數
a = int(input())
b = int(input())
max_range = a * b

for i in range(1, max_range+1): # 0 ~ max_range-1 -> 1 ~ max_range
  if i % a == 0 and i % b == 0:
    print(i)
    break
```

```python
# continue：迴圈中途暫停不往下執行，直接跳回迴圈起始處繼續執行

# 跳過數字：因為4不吉利，所以很多飯店都沒有四樓，可以根據使用者數入的數字，列出有哪些樓層
n = int(input())
for i in range(1, n+1): # 0 ~ n-1 -> 1 ~ n
  if i == 4:
    continue
  print(i)
```
